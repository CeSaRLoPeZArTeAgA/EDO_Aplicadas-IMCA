# Metodo de Euler Explicito para EDO

Sea el problema de valor inicial(PVI):

$$   y=f(t,y) $$
con
$y(a)=\alpha$  y  $a<t<b$


Donde $f$ una funcion Lipschitziana en la variable $y$. El método consiste en lo siguiente iteracion:

$$w_{i+1}=w_{i}+hf(t,w_{i})$$

con:


* Debe existir seguridad sobre la continuidad de la función f en el intervalo [a,b]



```python
#EULER EXPLICITO - EDO
import numpy as np
import matplotlib.pyplot as plt

#definicion de la funcion f
def f(t,y):
  return y-t**2+1
#Metodo de Euler
def Euler(a,b,y0,f,N):
  """
  Implementacion metodo de Euler Explicito
  a--Inicio Intervalo
  b--Fin de Intervalo
  y0--Aproximacion Inicial
  f-funcion
  N--Pasos
  """
  h=(b-a)/N
  t=a
  w=y0
  Y,T=[],[]
  for i in range(1,N+1):
    w=w+h*f(t,w)
    Y+=[w]
    t=a+i*h
    T+=[t]
  return T,Y
```

```python
# Definicion de la funcion solucion Y
def ff(T):
  y_real=[]
  for t in T:
    y_real+=[(t+1)**2-0.5*np.exp(t)]
  return y_real;
```

```python
#Euler para h=0.5
tr,yr1=Euler(0,5,0.5,f,10)
yv=ff(tr)
plt.plot(tr,yr1)
plt.plot(tr,yv)
plt.grid(True)
plt.xlabel('t')
plt.ylabel('y')
plt.title('Metodo de Euler  Explicito - h')
plt.legend(['Euler-Expl h','Function Real'])
plt.show()
```

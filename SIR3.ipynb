```python
# Carga de Librerias
import numpy as np
import matplotlib.pyplot as plt
```

```python
################################################
# Parámetros del modelo SIR - model 1
alpha = 0.4998   # Tasa de infección
lamda = 0.0833  # Tasa de recuperación

# Condiciones iniciales
S0 = 1  # Población susceptible inicial
I0 = 0.001  # Población infectada inicial
R0 = 0   # Población recuperada inicial

# Paso de tiempo
h = 0.1

# Tiempo final
t_final = 120

# Número de pasos
n_steps = int(t_final / h)

# Inicializamos los vectores de tiempo y solución
t_values = np.linspace(0, t_final, n_steps + 1)

# Para almacenar S(t), I(t) y R(t)
y_values = np.zeros((n_steps + 1, 3))

# Establecemos las condiciones iniciales
y_values[0] = [S0, I0, R0]

# Método de Euler
for i in range(n_steps):
  S, I, R = y_values[i]
  dydt = np.array([-alpha * S * I,
                    alpha * S * I - lamda * I,
                    lamda * I])
  y_values[i + 1] = y_values[i] + h * dydt

# Creación de DataFrame para mostrar los resultados
import pandas as pd
results_df = pd.DataFrame({
  't': t_values,
  'S(t)': y_values[:, 0],
  'I(t)': y_values[:, 1],
  'R(t)': y_values[:, 2]
})

# Mostrar un fragmento del cuadro de resultados
print("Presentacion de los primero 15 resultados:")
print(results_df.head(15))

# Graficamos los resultados
plt.figure(figsize=(10, 6))
plt.plot(t_values, y_values[:, 0], label='S(t) - Susceptibles')
plt.plot(t_values, y_values[:, 1], label='I(t) - Infectados')
plt.plot(t_values, y_values[:, 2], label='R(t) - Recuperados')
plt.title("ESCENEARIO 01:Tasa Transmision "+r'$\alpha= 0.4998 $'+
          ", Tasa de recuperacion "+ r'$\lambda=0.0833$'+
          "\n Tasa Basica de resproducion "+r'$R_{0}=6$')
plt.xlabel('Tiempo (dias)')
plt.ylabel('Numero de Personas '+r'$(x10^{3})$')
plt.legend()
plt.grid(True)
plt.savefig('escenario_1.jpg', format='jpg')
plt.show()
```
